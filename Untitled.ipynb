# ë‚ ì§œ, ì‹œê°„ê°’ sin,cos ë³€í™˜

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import copy
from datetime import date
from sklearn.preprocessing import FunctionTransformer

def sin_transformer(period):
    return FunctionTransformer(lambda x: np.sin(x / period * 2 * np.pi))
  
def cos_transformer(period):
    return FunctionTransformer(lambda x: np.cos(x / period * 2 * np.pi))

X_2["month_sin"] = sin_transformer(12).fit_transform(X_2[["month"]])
X_2["month_cos"] = cos_transformer(12).fit_transform(X_2[["month"]])

X_2["day_sin"] = sin_transformer(30).fit_transform(X_2[["day"]])
X_2["day_cos"] = cos_transformer(30).fit_transform(X_2[["day"]])

X_2["hour_sin"] = sin_transformer(24).fit_transform(X_2[["hour"]])
X_2["hour_cos"] = cos_transformer(24).fit_transform(X_2[["hour"]])

# X_2["winddirection_sin"] = sin_transformer(360).fit_transform(X_2[["winddirection"]])
# X_2["winddirection_cos"] = cos_transformer(360).fit_transform(X_2[["winddirection"]])

fig, ax = plt.subplots(2, 1, sharex=True, figsize=(30,8))
X_2[["month_sin", "month_cos"]].plot(ax=ax[0])
X_2[["day_sin", "day_cos"]].plot(ax=ax[1])
# X_2[["winddirection_sin", "winddirection_cos"]].plot(ax=ax[2])
plt.suptitle("Cyclical encoding with sine/cosine transformation");
```

```python
from sklearn.model_selection import train_test_split
x_data=X_2.loc[:, ['temperature','windspeed','winddirection','humidity','month_sin','month_cos','day_sin','day_cos','hour_sin','hour_cos']]
y_data=X_2.loc[:, 'pressure']
x_train, x_test, y_train, y_test=train_test_split(x_data,
                                                  y_data,
                                                  test_size=0.2,
                                                  shuffle=True,
                                                  random_state=12)
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
#ì„ í˜• íšŒê·€ ëª¨í˜•
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train, y_train)

print("íšŒê·€ê³„ìˆ˜(ê¸°ìš¸ê¸°):", np.round(lr.coef_, 1)) #np: numpy, coef_: í”¼ì²˜ì— ëŒ€í•œ íšŒê·€ ê³„ìˆ˜ ê°’
print("ìƒìˆ˜í•­(ì ˆíŽ¸):", np.round(lr.intercept_, 1))  #intercept_: ìƒìˆ˜í•­(ì ˆíŽ¸) ê°’
```

```python
#ì„±ëŠ¥ í‰ê°€
from sklearn.metrics import mean_squared_error
y_train_pred=lr.predict(x_train)

train_mse=mean_squared_error(y_train, y_train_pred) #í›ˆë ¨ ë°ì´í„°ì˜ í‰ê°€ ì ìˆ˜
print("Train MSE:%.4f" % train_mse)

test_mse=mean_squared_error(y_test, y_test_pred)
print("Test MSE:%.4f" % test_mse)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_data,lr.predict(x_data))
print("R2:%.4f" % r2)
```
